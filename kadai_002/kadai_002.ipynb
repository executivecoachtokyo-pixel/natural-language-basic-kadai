```python
#Step
#Step1：スクレイピング
#Step2：HTMLタグ・改行などの削除
#Step3：最初の1文を手動で品詞分割
#Step4：ストップワード除去


#Step1. スクレイピング

import requests
from bs4 import BeautifulSoup
import re

#青空文庫『文壇の趨勢』のURL
url = "https://www.aozora.gr.jp/cards/000148/files/2371_13943.html"

#HTMLを取得
response = requests.get(url)

#文字化け対策
response.encoding = response.apparent_encoding

#BeautifulSoupでHTMLを解析
soup = BeautifulSoup(response.text, "html.parser")

#本文部分を取得
main_text = soup.find("div", class_="main_text")

#本文の中身を表示
print("取得した本文の一部:")
print(main_text.text[:500])
```

```python
#Step2. HTMLタグや不要な文字列の削除

#.textでHTMLタグを除去した本文を取得
text = main_text.text

#改行・空白を削除
clean_text = re.sub(r"\s+", "", text)

print("前処理後の本文の一部:")
print(clean_text[:500])
```

```python
#Step3. ストップワードの除去
#対象文：
#「近頃は文壇などというものは段々なくなってしまうように思っています。」

#最初の1文を手動で品詞ごとに区切る
tokens = [
    {"word": "近頃", "part": "名詞"},
    {"word": "は", "part": "助詞"},
    {"word": "文壇", "part": "名詞"},
    {"word": "など", "part": "助詞"},
    {"word": "という", "part": "連語"},
    {"word": "もの", "part": "名詞"},
    {"word": "は", "part": "助詞"},
    {"word": "段々", "part": "副詞"},
    {"word": "なくなって", "part": "動詞"},
    {"word": "しまう", "part": "補助動詞"},
    {"word": "ように", "part": "助動詞的表現"},
    {"word": "思って", "part": "動詞"},
    {"word": "います", "part": "補助動詞"}
]

#ストップワードを設定する
stop_words = [
    "は",
    "など",
    "という",
    "もの",
    "しまう",
    "ように",
    "います"
]

#ストップワードを除去する
filtered_tokens = []

for token in tokens:
    if token["word"] not in stop_words:
        filtered_tokens.append(token)

#結果を表示する
print("手動で品詞ごとに区切った結果:")
for token in tokens:
    print(token)

print("\nストップワード:")
print(stop_words)

print("\nストップワード除去後:")
for token in filtered_tokens:
    print(token)
```
